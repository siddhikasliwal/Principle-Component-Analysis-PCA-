#**Linear Regression**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Marketing_Data-Linear.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
#splitting into dependent and independent variable
x = df.iloc[:,:-1]
y = df['sales']
```

```python
x.head()
```

```python
y.head()
```

```python
x.shape, y.shape
```

```python
#splitting into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, train_size = 0.33, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
X_train = sc_x.fit_transform(X_train)
X_test  = sc_x.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
#train the model using the training sets
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print("Accuracy" , r2_score(y_test , y_pred)*100,"%")
```

**After PCA**

```python
from sklearn.decomposition import PCA
```

the fit() method will allow us to get the parameters of the scaling function. The transform() method will transform the dataset to proceed with further data analysis steps. The fit_transform() method will determine the parameters and transform the dataset.

```python
pca = PCA(n_components=2)
X_train = pca.fit_transform(X_train,y_train)
X_test = pca.transform(X_test)
```

```python
reg = LinearRegression()
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
```

```python
print("Accuracy" , r2_score(y_test , y_pred)*100,"%")
```

***After applying PCA to Regression model the acuraccy is decreased***

*   Before PCA - Accuracy 83.85 %
*   After PCA  - Accuracy 78.23 %

#**Logistic regression**

```python
df=pd.read_csv("/content/Amazon Sale Report.csv")
df.head()
```

```python
df.drop(['index','Order ID','Date','Style','SKU','ASIN','ship-postal-code','promotion-ids','ship-city','B2B','Unnamed: 22'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df[(df['Status'] == 'Cancelled') | (df['Status'] == 'Shipped') | (df['Status'] == 'Shipped - Delivered to Buyer') | (df['Status'] == 'Shipped - Out for Delivery') | (df['Status'] == 'Pending') | (df['Status'] == 'Shipping')]
df["Status"] = df.Status.map({"Cancelled":0, "Shipped" : 1, "Shipped - Delivered to Buyer":2, "Shipped - Out for Delivery":3, "Pending":4, "Shipping":5})
```

```python
df.rename(columns = {'Courier Status': 'Courier_Status'}, inplace = True) 
```

```python
df = df[(df['Courier_Status'] == 'Shipped') | (df['Courier_Status'] == 'Unshipped')]
df['Courier_Status'] = df.Courier_Status.map({"Shipped":1, "Unshipped":0})
```

```python
df["Fulfilment"] = df.Fulfilment.map({"Merchant":0, "Amazon" : 1})
```

```python
df.rename(columns = {'Sales Channel ': 'Sales_Channel'}, inplace = True) 
df['Sales_Channel'] = df.Sales_Channel.map({"Amazon.in":1, "Non-Amazon":0})
```

```python
df.rename(columns = {'ship-service-level': 'ship_service_level'}, inplace = True) 
df['ship_service_level'] = df.ship_service_level.map({"Standard":1, "Expedited":0})
```

```python
df["Category"] = df.Category.map({"kurta":0, "Top":1, "Set":2, "Western Dress":3, "Ethnic Dress":4, "Saree":5, "Blouse":6, "Bottom":7, "Dupatta":8})
```

```python
df["Size"] = df.Size.map({"XS":0, "S":1, "M":2, "L":3, "XL":4, "XXL":5, "3XL":6, "4XL":7, "5XL":8, "6XL":9, "Free":10 })
```

```python
df['currency'] = df.currency.map({"INR":1, "nan":0})
df['currency'].fillna('0')
```

```python
df['currency'].fillna(value = 0, inplace = True)
df['Amount'].fillna(value = 0, inplace = True)
```

```python
df.rename(columns= {"fulfilled-by" : "fulfilled_by"},inplace = True)
df['fulfilled_by'] = df['fulfilled_by'].fillna('Difficult Ship')
df["fulfilled_by"] = df.fulfilled_by.map({"Easy Ship":1, "Difficult Ship":0})
```

```python
df1 = df.loc[:,['Courier_Status','Status','Fulfilment','Sales_Channel','ship_service_level','Category','Size','Qty','ship_country','fulfilled_by','currency','Amount']]
```

```python
df1.isnull().sum()
```

```python
df1.drop(['fulfilled_by'],inplace = True, axis = 1)
```

```python
df1.head()
```

```python
#Split dataset into independent and dependent features
X = df1.iloc[:,2:]
y = df1.iloc[:,1]
```

```python
#splitting the dataset into the training and testing set
from sklearn.model_selection import train_test_split
X_Train, X_Test, y_Train, y_Test = train_test_split(X,y,test_size = 0.25, random_state = 0)
X_Train.shape, y_Train.shape, X_Test.shape, y_Test.shape
```

```python
#Scaling the Data
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_Train = sc_X.fit_transform(X_Train)
X_Test = sc_X.transform(X_Test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_Train,y_Train)
```

```python
y_Pred = classifier.predict(X_Test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_Test, y_Pred)
print("Accuracy", acc*100, "%")
```

**After PCA**

```python
pca = PCA(n_components=2)
X_train = pca.fit_transform(X_Train,y_Train)
X_test = pca.transform(X_Test)
```

```python
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_Train, y_Train)
```
